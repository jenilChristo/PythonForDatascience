Rank of a matrix

```python
a = np.array([[6,1,1],
              [4,-2,5],
              [2,8, 7]])
# number of independent row and columns

np.linalg.matrix_rank(a)
```

```python
b = np.array([
    [1,2,3],
    [2,3,4],
    [3,5,7]
])
np.linalg.matrix_rank(b)
```
