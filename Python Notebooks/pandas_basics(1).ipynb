```python
import seaborn as sns

sns.boxplot(x= 'Pclass', y= 'Age', data = titanic_df)
```

```python
def age_imputuation(cols):
   Age = cols[0]
   Pclass = cols[1] 

   if pd.isnull(Age): 
     if Pclass==1:
        return 38
     elif Pclass ==2:
        return 35 
     else: return 30 
   else: return Age
```

```python
titanic_df['Age'] = titanic_df[['Age','Pclass']].apply(age_imputuation,axis=1)
```
