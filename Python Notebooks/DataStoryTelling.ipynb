```python
# cleaning the runs column
def cleanRuns(x):
  if '*' in str(x):
    return int(str(x.replace('*', '')))
  elif 'TDNB' in str(x) or 'DNB' in str(x) :
    return 0
  else :
    return int(x)

tendulkar['Runs'] = tendulkar['Runs'].apply(cleanRuns)
tendulkar['4s'] = tendulkar['4s'].apply(lambda x: 0 if '-' in str(x) else int(x))

tendulkar['4s'].unique()
```

```python
0-10 runs

10-20 runs

20-30 runs

30-40 runs
```

```python
def runsCategory(x):
  if x in range(0, 11):
    return '0-10 runs'
  elif x in range(11, 21):
    return '10-20 runs'
  elif x in range(21,30):
    return '20-30 runs'
  elif x in range(31, 40):
    return '30-40 runs'
  else:
    return '> 40 runs'
   
tendulkar['runs_category'] = tendulkar['Runs'].apply(runsCategory)

tendulkar.head()
```
