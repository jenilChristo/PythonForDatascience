```python

C = [2, 5, 9, 12, 13, 15, 16, 17, 18, 19]
F = [2, 4, 5, 6, 7, 9, 13, 16]
H = [1, 2, 5, 9, 10, 11, 12, 13, 15]
cricket_set = set(C)
football_set = set(F)
hockey_set = set(H)

play_all = list(cricket_set.intersection(football_set).intersection(hockey_set))
play_all.sort()

cricket_and_football = list(cricket_set.intersection(football_set).difference(hockey_set))
cricket_and_football.sort()

exact_two_sport = {}

for i in cricket_set:

  if(i not in exact_two_sport.keys() ):
    exact_two_sport[i] = 1
  else:
    exact_two_sport[i] += 1

for i in football_set:
  
  if(i not in exact_two_sport.keys() ):
    exact_two_sport[i] = 1
  else:
    exact_two_sport[i] += 1

for i in hockey_set:
  
  if(i not in exact_two_sport.keys() ):
    exact_two_sport[i] = 1
  else:
    exact_two_sport[i] += 1

def filterByValue(pair):
  key,value = pair
  if value == 2:
    return True
  else:
    return False

filtered_dict = dict(filter(filterByValue, exact_two_sport.items()))
answer_3 = list(filtered_dict.keys())
answer_3.sort()


not_play_any = cricket_set.union(football_set).union(hockey_set)

answer_4 = []

for i in range(1,21):
  if(i not in not_play_any):
    answer_4.append(i)

answer_4.sort()



print(answer_4)

# answer_1 = list(map(lambda x, y : x if x < y else y, play_all))
# print(exact_two_sport)

# (Make sure your lists are sorted in ascending order before you print them out)
```
